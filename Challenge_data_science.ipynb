## **Importando los Datos:**

```python
import pandas as pd
import matplotlib.pyplot as plt

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
```

```python
las_tiendas = [tienda1+tienda2+tienda3+tienda4]
```

## **1)Análisis de la facturación**

```python
#Calculando el ingreso total de cada tienda

total_tienda1=tienda1["Precio"].sum()
total_tienda2=tienda2["Precio"].sum()
total_tienda3=tienda3["Precio"].sum()
total_tienda4=tienda4["Precio"].sum()

print(f'El ingreso total de la tienda 1 es ${total_tienda1:.0f}')
print(f'El ingreso total de la tienda 2 es ${total_tienda2:.0f}')
print(f'El ingreso total de la tienda 3 es ${total_tienda3:.0f}')
print(f'El ingreso total de la tienda 4 es ${total_tienda4:.0f}')
```

### La tienda con más ingresos es la **tienda 1**

## **2)Ventas por categoría**

```python
#Cantidad de productos vendidos por categoría en cada tienda


productos1=tienda1["Categoría del Producto"].value_counts().sort_values(ascending=False)
productos2=tienda2["Categoría del Producto"].value_counts().sort_values(ascending=False)
productos3=tienda3["Categoría del Producto"].value_counts().sort_values(ascending=False)
productos4=tienda4["Categoría del Producto"].value_counts().sort_values(ascending=False)


print("Tienda 1\n",productos1,"\n")
print("Tienda 2\n",productos2,"\n")
print("Tienda 3\n",productos3,"\n")
print("Tienda 4\n",productos4,"\n")
```

###Las categorías con más ventas fueron **Muebles** y **Electrónicos**

## **3)Calificaciones promedio de cada tienda**

```python
#Promedio de la calificación de los clientes para cada tienda

calificacion1=tienda1["Calificación"].mean()
calificacion2=tienda2["Calificación"].mean()
calificacion3=tienda3["Calificación"].mean()
calificacion4=tienda4["Calificación"].mean()

print(f'La calificación promedio de la tienda 1 es: {calificacion1:.2f}')
print(f'La calificación promedio de la tienda 2 es: {calificacion2:.2f}')
print(f'La calificación promedio de la tienda 3 es: {calificacion3:.2f}')
print(f'La calificación promedio de la tienda 4 es: {calificacion4:.2f}')
```

### La mejor clasificación la tiene la **tienda 3**

## **4)Productos con mayor y menor ventas**

```python
mas_ventas1 = tienda1.groupby("Producto")["Producto"].count().sort_values(ascending=False)
mas_ventas2 = tienda2.groupby("Producto")["Producto"].count().sort_values(ascending=False)
mas_ventas3 = tienda3.groupby("Producto")["Producto"].count().sort_values(ascending=False)
mas_ventas4 = tienda4.groupby("Producto")["Producto"].count().sort_values(ascending=False)

print("Los más vendidos de la tienda 1 son:")
print(mas_ventas1.head(2), "\n")
print("Los menos vendidos de la tienda 1 son:")
print(mas_ventas1.tail(2), "\n")

print("--------------")
print("Los más vendidos de la tienda 2 son:")
print(mas_ventas2.head(2), "\n")
print("Los menos vendidos de la tienda 2 son:")
print(mas_ventas2.tail(2), "\n")

print("--------------")
print("Los más vendidos de la tienda 3 son:")
print(mas_ventas3.head(2), "\n")
print("Los menos vendidos de la tienda 3 son:")
print(mas_ventas3.tail(2), "\n")

print("--------------")
print("Los más vendidos de la tienda 4 son:")
print(mas_ventas4.head(2), "\n")
print("Los menos vendidos de la tienda 4 son:")
print(mas_ventas4.tail(2), "\n")
```

## **5)Costo de envío promedio de cada tienda**

```python
costo_envio1 = tienda1['Costo de envío'].mean()
costo_envio2 = tienda2['Costo de envío'].mean()
costo_envio3 = tienda3['Costo de envío'].mean()
costo_envio4 = tienda4['Costo de envío'].mean()

print(f" El valor total de envíos de la tienda 1 es ${costo_envio1:.2f}")
print(f" El valor total de envíos de la tienda 2 es ${costo_envio2:.2f}")
print(f" El valor total de envíos de la tienda 3 es ${costo_envio3:.2f}")
print(f" El valor total de envíos de la tienda 4 es ${costo_envio4:.2f}")
```

### El menor costo de envío lo tiene la **tienda 4**

## **- Gráficos**

### **Ingresos de las tiendas**

```python
ingresos_tienda1 = 1150880400.0
ingresos_tienda2 = 1116343500.0
ingresos_tienda3 = 1098019600.0
ingresos_tienda4 = 1038375700.0

tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
totales = [ingresos_tienda1, ingresos_tienda2, ingresos_tienda3, ingresos_tienda4]

plt.figure(figsize=(7, 5))
plt.bar(tiendas, totales)

plt.title("Valor total de ventas por tienda")
plt.xlabel("Tienda")
plt.ylabel("Suma total (Precio)")
plt.grid(axis="y", linestyle="--", alpha=0.5)


for i, valor in enumerate(totales):
    plt.text(i, valor, f"{valor:,.0f}", ha='center', va='bottom')

plt.show()
```

### **Ventas según la categoría**

```python
total_categorias = (
    productos1
    .add(productos2, fill_value=0)
    .add(productos3, fill_value=0)
    .add(productos3, fill_value=0)
)

total_categorias = total_categorias.sort_values(ascending=True)

plt.figure(figsize=(10, 10))
plt.pie(
    total_categorias.values,
    labels=total_categorias.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Cantidad total de ventas de todas las tiendas")
plt.axis("equal")
plt.show()
```
